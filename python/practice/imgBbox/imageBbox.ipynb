```python
import numpy as np
import pandas as pd
import json
import cv2
import matplotlib.pyplot as plt
```

```python
with open("./V001_tom1_00_052_a2_18_20220101_27_00092524_40158887.json","rb") as f:
    data = json.load(f)
    
shapes = data["shapes"]
shape = shapes[0]
point = shape["points"]
point = list(map(lambda x: list(map(int,x)),point))
point
```

```python
pd.DataFrame({"a":[1],"b":[2]})
```

```python
img = cv2.imread('./V001_tom1_00_052_a2_18_20220101_27_00092524_40158887.png')
for shape in shapes:
    point = shape["points"]
    point = list(map(lambda x: list(map(int,x)),point))
    img = cv2.rectangle(img,(int(point[0][0]),int(point[0][1])),(int(point[1][0]),int(point[1][1])),(0, 255, 0),thickness=10)
print(img.shape)
cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
img.shape
```

```python
img2 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.axis('off') #  x y 
plt.imshow(img2)
plt.show()
```
